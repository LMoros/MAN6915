```python
import math
from collections import OrderedDict as odic

def problem2():
    def loaded(path):
        """
        Loads values from file
        """
        def getFields(line):
            """ This method strips the string "Iris-" out of all the fields in the lines
                (fields are separated by ',')
            """ 
            return [field.strip("\n").replace("Iris-","") for field in line.split(",")]        
        
        # Obtaining header and values from file
        with open(path, "r") as file:
            lines = file.readlines()
            headers = [field for field in getFields(lines[0])]
            values = [dict(zip(headers, getFields(line)))\
                           for line in lines[1:]]
        return headers, values            
        
    def processed(headers, values):
        """
        Calculates statistics for each fled of each species
        """
        def processField(values, header):
            """
            Calculating statiscs of a field
            """
            def count(values):
                """
                Counts how many items in 'Values' list
                """
                result = 0
                for value in values:
                    result+=1
                return result
            def add(values):
                """
                Adds the each value in the "values" list
                """
                result = 0
                for value in values:
                    result+=value
                return result
            def minimum(values):
                """
                Obtains the minimum value in the "Values" list
                """
                result = values[0]
                for value in values:
                    result = value if value<result else result
                return result
            def maximum(values):
                """
                Obtains the maximum value in the "values" list
                """
                result = values[0]
                for value in values:
                    result = value if value>result else result
                return result
            def mean(values):
                """
                Obtains the mean of the values
                """
                return add(values)/count(values)
            def variance(values):
                """
                Obatins the variance of the values
                """
                cnt = count(values)
                mn = mean(values)
                return add(map(lambda v : (v-mn)**2/cnt, values))
            def std(values):
                """
                Obtains the standard deviations of the values
                """
                return math.sqrt(variance(values))            
            
            return odic([("cnt",count(values))\
                    ,("sum",add(values))\
                    ,("min",minimum(values))\
                    ,("max",maximum(values))\
                    ,("mean",mean(values))\
                    ,("var",variance(values))\
                    ,("std",std(values))\
                    ,("header",header)])

        def processSp(sp,headers,values):
            """
            Processing one species
            """
            return [ processField([float(value[h]) for value in values if value["Species"]==sp],h)\
                     for h in headers if h != "Species" ] 
        #dedup species
        species = set([x["Species"] for x in values])
        
        #tiggering processing each species
        return [{sp:processSp(sp, headers, values)} for sp in species] 

    def output(x):
        """
        Displays a grid with the calulcated staticstics
        """
        def show(species, stats):
            """
            Prints the calulcated statistics
            """

            def formatStatLine(statName, statResult):
                """
                This function formats line showing a statics type
                """
                return '{:9} {:6.2f} '.format("{}:".format(statName), statResult)

            def getHeaders():
                """
                Obtains the headers
                """
                return [x['header'] for x in stats]

            def formatHeaders():
                """
                This function formats header line
                """
                return "{:11}{}".format(" ","".join(["|{:17}".format(h) for h in getHeaders()]))

            def formatSpecies():
                """
                This function formats the line that shows the species
                """
                return "{}{}".format(
                            "{:11}".format(species),\
                            # after the species name, show a deviding line (i.e. '======')
                            "{}".format("".join(["|{}".format("="*17) for h in getHeaders()])))

            def formatStats():
                """
                This function formats all the stats of a species
                """
                #obtain names of statistics
                statNames = []
                [statNames.append(key) for stat in stats for key in stat.keys() \
                     if key != 'header' and key not in statNames]
                return "\n".join(\
                          ["{:11}|{}".format(" ",\
                            "|".join(\
                            [formatStatLine(key,stat[key]) for stat in stats]\
                            )\
                           )\
                          for key in statNames]\
                        )

            #Print headers
            print(formatHeaders())

            #Print species and deviding line
            print(formatSpecies())

            #Print statistics
            print(formatStats())

            # add an empty line between Species
            print()        
        
        #Triggering showing the calculated statistics
        [show(sp, speciesDicionary[sp])\
             for speciesDicionary in x\
             for sp in speciesDicionary.keys()]

    #triggering loading, processing, and displaying    
    output(processed(*loaded("iris.txt")))
problem2()
```

```python
from collections import Counter 
import math
import string

def problem3():
    
    def removeNonAlpha(word):
        """ This methods removes non alphanumeric characters 
            from the string parameter 'words'
        """    
        return "".join([c for c in word if c.isalpha()])
    
    def loaded(path):
        """ Loads txt from file specified by parameter path
        """
        # loading file
        with open(path) as file:
            # get words from lines and filter non alphanumeric
            words = [removeNonAlpha(word) for line in file for word in line.split(" ")] 
            # return words that are not empty strings
            return [word for word in words if word != ""]
    
    def processed(words):
        """ This methods calculates the 10 most frequent words, 10 longest wrods
            mean word lengths, and word lenghts standard deviation
        """
        
        # obtain 10 mos frequent words
        freqWords = [(word,freq) for (word,freq) in Counter(words).most_common(10)]
        
        # deduplicate words before checking lengths and stats
        dedupWords = set(words)
        
        # obtain 10 longest words and thier lengths
        longestWords = [(word,len(word)) for word in\
                        sorted(dedupWords,key=lambda x : len(x), reverse=True)]\
                        [:10]
            
        # calculate mean        
        lens = [len(word) for word in dedupWords]
        mean = sum(lens)/len(dedupWords)
        
        # calculate standard deviation
        std = math.sqrt(sum(map(lambda t : (t-mean)**2/len(lens), lens)))
        
        #return tuple with results
        return freqWords, longestWords, mean, std 
    
    def output(frq,lng,mn,std):
        """ This method format and presents the 10 most frequestn words, 10 longest words, 
            word length mean, and word length standrad deviation
        """
        #unpack tuple
        # , std = frq_lng_mn_std
        
        # get max word length
        wordLen = max([length for word, length in lng ])+2

        # format and print words freqs
        print("Most Common Words:")
        print("")
        print("")
        print("{w:<{l}}|{c}"\
                         .format(w="Word"\
                                 , l=wordLen\
                                 , c="Freq"))
        print("..................................")
        print("\n".join(["{w:<{l}}|{c}"\
                         .format(w="{}:".format(x)\
                                 , l=wordLen\
                                 , c=y)\
                         for x,y in frq]))
        
        print("..................................")
        print()
        print()
        
         # format and print longest words
        print("Longest Words:")
        print()
        print("{w:<{l}}|{c}"\
                         .format(w="Word"\
                                 , l=wordLen\
                                 , c="Length"))
        print("..................................")       
        print("\n".join(["{w:<{l}}|{c}"\
                         .format(w="{}:".format(x), l=wordLen, c=y)\
                         for x,y in lng]))
        
        # format and print Mean
        print("..................................")
        print("Mean Word Length:     {:.2f}".format(mn))
        
        # format and print std
        print("..................................")
        print("Word Length Std. Dev: {:.2f}".format(std))
        
    output(*processed(loaded("bible.txt")))
    
problem3()
```
